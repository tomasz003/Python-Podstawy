## 2. ANN - sztuczne sieci neuronowe

Systemy przeznaczone do przetwarzania informacji inspirowane funkcjonalnością (nie budową!!!) biologicznego mózgu.

Bardzo różne architektury i sposoby uczenia ale praktycznie wszystkie wykorzystują:
- **warstwy perceptronów**: zazwyczaj gęste warstwy neuronów z nieliniowymi funkcjami aktywacji,
- **backproagation** (metoda wstecznej propagacji błędu): trenowanie sieci
- **gradient descent** (metoda najszybszego spadku): metoda optymalizacji

Żadna z tych metod/struktur nie występuje wprost w mózgu człowieka - dla przykładu backpropagation działa globalnie i wymaga przesyłania wiadomości (błędu) wstecz - od wyjścia do wejścia. Takich zjawisk w biologicznych mózgach się nie obserwuje.

### 3.4 Aktywacja

Każdy neuron posiada tzw. funkcję aktywacji. Dotychczas była to funkcja liniowa, w konsekwencji mamy układ liniowy. W takim systemie jesteśmy w stanie uzyskać tylko liniową konwersję danych wejściowych. Aby rozszerzyć możliwości sieci neuronowych, kluczowym pojęciem jest nieliniowa funkcja aktywacji $φ$.



$$
y = 𝜑 \left( b + \sum_{j=1}^M w_j x_j \right)
$$

```
# helper functions
import matplotlib.pyplot as plt
import numpy as np

def plot_activation(x, y, title_str):
    plt.plot(x, y)
    plt.xlabel('Neuron Input')
    plt.ylabel('Neuron Output')
    plt.title(title_str)
    plt.grid()
    plt.show()
```

```
#@title RELU (Rectified Linear Unit)
x = np.linspace(-5, 5, 100)
y = np.maximum(0, x)

plot_activation(x, y, 'ReLU, f(x) = max(0, x)')
```

```
#@title Sigmoid
x = np.linspace(-5, 5, 100)
y = y = 1/(1 + np.exp(-x))

plot_activation(x, y, 'Sigmoid, f(x) = 1/(1+exp(-x))')
```

## 4. Implementacja MLP (Circle Dataset)

Poniżej przedstawiono model w korzystający z API Keras realizujący dokładnie tą samą funkcjonalność jaka była w playground.
- dwa wejścia numeryczne
- jedna warstwa ukryta, trzy jednostki, w pełni połączone, funkcja aktywacji ReLU
- jedno wyjście

simple_MLP_small.png

```
#@title circes dataset
import numpy as np
from sklearn.model_selection import train_test_split

def gen_circles(N):
    N2 = int(N/2)
    # r_p = np.random.normal(0, 1, N2)
    # r_n = np.random.normal(5, 1, N2)
    r_p = np.random.uniform(0.0, 0.4, N2)
    r_n = np.random.uniform(0.6, 0.8, N2)
    angle_p = np.random.uniform(0, 2 * np.pi, N2)
    angle_n = np.random.uniform(0, 2 * np.pi, N2)
    X_p = np.stack((r_p*np.sin(angle_p), r_p*np.cos(angle_p)), axis=1)
    X_n = np.stack((r_n*np.sin(angle_n), r_n*np.cos(angle_n)), axis=1)
    y_p = np.ones((N2, 1))
    y_n = np.zeros((N2, 1))

    return np.concatenate((X_p, X_n)), np.concatenate((y_p, y_n))

# generate and plot
N = 2000
X, y = gen_circles(N)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, shuffle = True)
```

```
#@title review and plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.scatter(X_train[:, 0], X_train[:, 1], s=60, c=y_train)
plt.show()
```

```
#@title build model and train
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential

# model = Sequential()
model = tf.keras.Sequential()
model.add(Dense(3, activation='relu', input_dim=2, name="L1"))
model.add(Dense(1, activation='sigmoid', name="L2"))

model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```
model.fit(X_train, y_train,
          epochs=20,
          batch_size=4,
          verbose=1)
```

```
xx, yy = np.meshgrid(np.arange(-1, 1, 0.01), np.arange(-1, 1, 0.01))
y_pred = model.predict(np.c_[xx.ravel(), yy.ravel()])
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.contour(xx, yy, np.round(y_pred.reshape(xx.shape)))
plt.scatter(X_train[:, 0], X_train[:, 1], s=60, c=y_train)
plt.show()
```
